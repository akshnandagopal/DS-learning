```python
import pandas as pd
import numpy as np
import re
```

```python
fair_deal_df = pd.read_csv('1937_m3_datasets_v1_jp0_zpnuw9a/FairDealCustomerData.csv', header=None)
fair_deal_df
```

```python
# Add column names

headers = ['Family name', 'Name', 'Is Blacklist']
fair_deal_df.columns = headers
fair_deal_df
```

```python
# Separate Title, First and Last name using regex

titles = []
first_name = []
last_name = []
for name in fair_deal_df['Name']:
    pattern = r'\s([A-Za-z]+\.)\s([A-Za-z]+)\s*(.*)'
    titles.append(re.sub(pattern, r'\1', name))
    first_name.append(re.sub(pattern, r'\2', name))
    last_name.append(re.sub(pattern, r'\3', name))

fair_deal_df['Title'] = titles
print(fair_deal_df['Title'].value_counts())
fair_deal_df['First Name'] = first_name
print(fair_deal_df['First Name'].value_counts())
fair_deal_df['Last Name'] = last_name
print(fair_deal_df['Last Name'].value_counts())
```

```python
# Create exception if customer not eligible

class CustomerNotAllowedException(Exception):
    def __init__(self, message = "Not Eligible"):  # default message initialization
        self.message = message
        super().__init__(self.message)

class Customer:
    def __init__(self,first_name,fair_deal_df):
        #self.title = title
        self.first_name = first_name
        #self.last_name = last_name
        self.fair_deal_df = fair_deal_df

    def createOrder(self,customer_name):
        if self.fair_deal_df[self.fair_deal_df['First Name']==customer_name]['Is Blacklist'].values[0] == 1:
            raise CustomerNotAllowedException(f"{customer_name} is not eligible")   # customer not allowed exception
        else:
            return True                

customer_name = input()
customer = Customer(customer_name,fair_deal_df)
try:
    if customer.createOrder(customer_name):
        print('Customer is eligilble')
except CustomerNotAllowedException as customer_not:
    print('Exception was raised:',customer_not)
```
