```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1 Extracting the data

sal_df = pd.read_csv('1937_m4_datasets_v1_5fa_ae16cor/SalaryGender.csv')
print('Shape:', sal_df.shape)
```

```python
sal_df.head(5)
```

```python
# Data cleaning

sal_df['Gender'] = sal_df['Gender'].replace({1:'M', 0:'F'})
sal_df['PhD'] = sal_df['PhD'].replace({1:'Yes', 0:'No'})
sal_df
```

```python
#1 Storing Df columns in numpy arrays

sal_array1 = np.array(sal_df['Salary'])
sal_array2 = np.array(sal_df['Gender'])
sal_array3 = np.array(sal_df['Age'])
sal_array4 = np.array(sal_df['PhD'])
print('Column1\n', sal_array1)
print('Column2\n', sal_array2)
print('Column3\n', sal_array3)
print('Column4\n', sal_array4)
```

```python
print(sal_df['Gender'].value_counts())
```

```python
#2

mphd = sal_df[(sal_df['Gender'] == 'M') & (sal_df['PhD'] == 'Yes')]
print('No. of Men with PhD:', mphd.shape[0])

wphd = sal_df[(sal_df['Gender'] == 'F') & (sal_df['PhD'] == 'Yes')]
print('No. of Women with PhD:', wphd.shape[0])
```

```python
#3 create a new df to store desired columns

columns = ['Age','PhD']
new_df = sal_df[columns]
print(new_df)

# another way
# new_df2 = sal_df[['Age','PhD']]
# print(new_df2)

# PhD value counts
print('Value counts before\n',new_df['PhD'].value_counts())

# delete data of all people who don't have a PhD

indices_to_drop = new_df[new_df['PhD'] == 'No'].index
new_df = new_df.drop(indices_to_drop)
print('New df with only PhD candidate data\n', new_df)

print('Value counts after\n',new_df['PhD'].value_counts())
```

```python
#4 total no. of people who have PhD

phd = sal_df[sal_df['PhD'] == 'Yes']
print('Total no. of people with PhD degree: ', phd.shape[0])
```

```python
#8 create 10x10 array with random values and find min and max

arr = np.random.randint(0,100,size=(10,10))
print('Array\n', arr)
print('Min:', arr.min())
print('Max:', arr.max())

# using rand - floating point values btn 0 and 1
arr1 = np.random.rand(10,10)
print('Array\n', arr1)
print('Min:', arr1.min())
print('Max:', arr1.max())

# can also be done using random.random with double for loop to iterate rows and columns or random((row,col))
# floating point values btn 0 and 1
row,col = 10,10
arr2 = [[np.random.random() for j in range(col)] for i in range(row)] # list comprehension
arr2 = np.array(arr2)
print('Array using random\n',arr2)
# to check the shape
print('Shape:', arr2.shape)
```

```python
#9 create a random vector of size 30 and find mean

arr = np.random.random(30)
print('Array\n', arr)
print('Mean\n', arr.mean())

# arr = np.array([np.random.random() for i in range(30)])
```

```python
# 10 numpy array containing 0-9 and negate all element between 3 and 9

arr = np.arange(0,10)
for i in range(len(arr)):
    if arr[i] in range(4,9):
        arr[i] = arr[i] * -1
print(arr)

# alternate way
arr1 = np.arange(10)
arr1[(arr1>3) & (arr1<9)] *= -1
print('Alternate way\n',arr1)

# boolean indexing 
print('Boolean indexing')
print(arr1[(arr1>3)|(arr1<2)])
print(arr1[(arr1>1) & (arr1<9)])
```

```python
#15 analyze school grade performance
#a  lit  review,  data  collection, descriptive and inferential statistics, and data visualization

# Data collection

school_data = pd.read_csv('1937_m4_datasets_v1_5fa_ae16cor/middle_tn_schools.csv')
school_data.head()
```

```python
# Identify req columns and Group by school ratings

req_cols = school_data[['reduced_lunch', 'school_rating']]
print(req_cols)

grouped_data = req_cols.groupby('school_rating').describe()
grouped_data
```

```python
# Correlation analysis

req_cols.corr()
```

```python
# Scatter plot

X, Y = req_cols['reduced_lunch'], req_cols['school_rating']
plt.scatter(X,Y)
plt.xlabel('Reduced Lunch')
plt.ylabel('School Rating')
plt.title('School Rating vs Reduced Lunch')
plt.show()
```

```python
# Correlation matrix

# columns of the df to compare against
filtered_cols = ['size','reduced_lunch','school_rating','stu_teach_ratio']
new_df = school_data[filtered_cols]
print(new_df)

correlation = new_df.corr()
corr_diagram = correlation.style.background_gradient(cmap='coolwarm')   # matshow,seaborn heatmap can also be used
corr_diagram
```
