```python
import pandas as pd
import numpy as np
```

```python
#1 Read the data 

# Data wrangling 
# Student data not revealed
# Relation of scores with age etc

math_df = pd.read_csv('1937_m4_datasets_v1_5fa_ae16cor/MathScoreTerm1.csv')
ds_df = pd.read_csv('1937_m4_datasets_v1_5fa_ae16cor/DSScoreTerm1.csv')
phy_df = pd.read_csv('1937_m4_datasets_v1_5fa_ae16cor/PhysicsScoreTerm1.csv')
print(math_df)
print(ds_df)
print(phy_df)
```

```python
print(math_df.shape)
print(ds_df.shape)
print(phy_df.shape)
print(math_df.head(5))
print(ds_df.head(5))
print(phy_df.head(5))
```

```python
#2 Removing name and ethnicity columns 

math_df.drop(['Name','Ethinicity'], axis = 1, inplace = True)
ds_df.drop(['Name','Ethinicity'], axis = 1, inplace = True)
phy_df.drop(['Name','Ethinicity'], axis = 1, inplace = True)
print('Math scores\n', math_df, '\nDS scores\n', ds_df, '\nPhysics scores', phy_df)
```

```python
#3 fill missing scores with zero for all 3 datasets

all_subject_scores = [math_df,ds_df,phy_df]
for df in all_subject_scores:
    # missing score data
    no_of_missing_scores = df['Score'].isnull().value_counts()
    print(no_of_missing_scores)

    # view the missing score entries
    missing_score = df[df['Score'].isnull()]
    print(missing_score)

    # fill missing score with zero

    df['Score'] = df['Score'].fillna(0)
    print('Missing values after update', df['Score'].isnull().sum())

# print all dataframes for validation
print('Math scores\n', math_df, '\nDS scores\n', ds_df, '\nPhysics scores\n', phy_df)

```

```python
# Renaming score columns and removing subject column for better merging  

math_df = math_df.rename(columns={'Score':'Math Score'})
math_df = math_df.drop('Subject',axis=1)
ds_df = ds_df.rename(columns={'Score':'DS Score'})
ds_df = ds_df.drop('Subject',axis=1)
phy_df = phy_df.rename(columns={'Score':'Physics Score'})
phy_df = phy_df.drop('Subject',axis=1)

print('Math scores\n', math_df, '\nDS scores\n', ds_df, '\nPhysics scores\n', phy_df)
```

```python
#4 merget all the three datasets

final_doc = pd.merge(pd.merge(math_df,ds_df,on=['ID','Age','Sex']), phy_df, on=['ID','Age','Sex'])
print(final_doc)
```

```python
# reorder columns 

col_order = ['ID', 'Age', 'Sex','Math Score','DS Score','Physics Score']
final_doc = final_doc[col_order]
print(final_doc)
```

```python
#5 change M/F to 1/2 for further analysis

#pd.set_option('future.no_silent_downcasting', True)

final_doc['Sex'] = final_doc['Sex'].replace({'M':1,'F':2})
print(final_doc)
```

```python
#6 Store resultant dataframe as a new csv file

final_doc.to_csv('ScoreFinal.csv')
```

```python
# replace value of ethnicity to numeric values

pd.set_option('future.no_silent_downcasting', True)

math_df2 = pd.read_csv('1937_m4_datasets_v1_5fa_ae16cor/MathScoreTerm1.csv')
ds_df2 = pd.read_csv('1937_m4_datasets_v1_5fa_ae16cor/DSScoreTerm1.csv')
phy_df2 = pd.read_csv('1937_m4_datasets_v1_5fa_ae16cor/PhysicsScoreTerm1.csv')
print(math_df2['Ethinicity'].value_counts())
print(ds_df2['Ethinicity'].value_counts())
print(phy_df2['Ethinicity'].value_counts())

# replacing in math df
math_df2['Ethinicity'] = math_df2['Ethinicity'].replace({'White American': 1, 'European American': 2, 'Hispanic': 3,'African American': 4})
print(math_df2['Ethinicity'].value_counts())
print(math_df2)
```

```python
# replacing in ds df
ds_df2['Ethinicity'] = ds_df2['Ethinicity'].replace({'White American': 1, 'European American': 2, 'Hispanic': 3,'African American': 4})
print(ds_df2['Ethinicity'].value_counts())
print(ds_df2)

# replacing in phy df
phy_df2['Ethinicity'] = phy_df2['Ethinicity'].replace({'White American': 1, 'European American': 2, 'Hispanic': 3,'African American': 4})
print(phy_df2['Ethinicity'].value_counts())
print(phy_df2)
```

```python
# fill na with mean

# math df2
print('Before na fill',math_df2.isna().sum())
mean = math_df2['Score'].mean()
print('Mean is', mean)
math_df2['Score'] = math_df2['Score'].fillna(mean)
print('After na fill',math_df2.isna().sum())
```

```python
# ds df2
print('Before na fill',ds_df2.isna().sum())
mean = ds_df2['Score'].mean()
ds_df2['Score'] = ds_df2['Score'].fillna(mean)
print('After na fill',ds_df2.isna().sum())

# phy df2
print('Before na fill',phy_df2.isna().sum())
mean = phy_df2['Score'].mean()
phy_df2['Score'] = phy_df2['Score'].fillna(mean)
print('After na fill',phy_df2.isna().sum())
```
