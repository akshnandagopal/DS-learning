```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1 Extracting Hurricanes data

h_df = pd.read_csv('1937_m4_datasets_v1_5fa_ae16cor/Hurricanes.csv')
h_df
```

```python
# Year vs no. of hurricanes

plt.bar(h_df['Year'],h_df['Hurricanes'])
plt.xlabel('Year')
plt.ylabel('No. of Hurricanes')
plt.title('No. of hurricanes yearwise analysis')
plt.show()
```

```python
#2 Extracting CityTemps data

ct_df = pd.read_csv('1937_m4_datasets_v1_5fa_ae16cor/CityTemps.csv')
ct_df
```

```python
ct_df['Year'].value_counts()
```

```python
# Temperatures by city histogram

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(ct_df['San Francisco'], bins=10, color='green', edgecolor='yellow')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('San Francisco')

plt.subplot(1,2,2)
plt.hist(ct_df['Moscow'], bins=10,color='blue', edgecolor='yellow')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Moscow')

# view histogram
plt.tight_layout()
plt.show()
```

```python
#4  simple plot 
# X = [1,2,3,4] y = [20, 21, 20.5, 20.8]

X = [1,2,3,4]
y = [20, 21, 20.5, 20.8]

plt.plot(X,y)
plt.show()
```

```python
#5

plt.plot(X,y, linestyle='--', marker='o', color='b')
plt.show()
```

```python
print(min(X),max(X),min(y),max(y))
```

```python
# axis configuration and titles,labels, error bar, figsize, dpi

y_error = [0.12, 0.13, 0.2, 0.1]

plt.figure(figsize=(4,5),dpi=100)
plt.errorbar(X, y, yerr=y_error, linestyle='--', marker='o', color='b')
plt.axis([min(X),max(X),min(y),max(y)])
plt.title('Simple Plot',fontsize=14)
plt.xlabel('X Axis',fontsize=14)
plt.ylabel('Y Axis',fontsize=14)
plt.show()
```

```python
# scatter plot of x and y 

X = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(5, 5))
plt.scatter(X, y)
plt.title('Scatter Plot of Random Values')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# create a dataframe

data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 
'female': [0, 1, 1, 0, 1],'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, 2, 3],
'postTestScore': [25, 94, 57, 62, 70]
}

df = pd.DataFrame(data)
df
```

```python
# scatter plot of pretestscore and posttestscore

plt.scatter(df['preTestScore'],df['postTestScore'],s=df['age'])
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.title('Scatter plot of preTestScore and postTestScore')
plt.show()
```

```python
# scatter plot with color

plt.scatter(df['preTestScore'],df['postTestScore'],s=300, c=df['female'], cmap='bwr',alpha=0.6)
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.title('Scatter plot of preTestScore and postTestScore')
plt.colorbar(label='Gender (0: Male, 1: Female)')
plt.show()
```
