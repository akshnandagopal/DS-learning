```python
#2 Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.

df['Total_Benefitted'] = df[['No. of Inmates benefitted by Elementary Education',
       'No. of Inmates benefitted by Adult Education',
       'No. of Inmates benefitted by Higher Education',
       'No. of Inmates benefitted by Computer Course']].sum(axis=1)
df

#df['Total_Benefitted'] = df['No. of Inmates benefitted by Elementary Education'] + df['No. of Inmates benefitted by Adult Education'] + df['No. of Inmates benefitted by Higher Education'] + df['No. of Inmates benefitted by Computer Course']
```

```python
# Create a new row -“totals” that is the sum of all inmates benefitted through each mode across all states.

df.loc[len(df)] = ['Total', '', df['No. of Inmates benefitted by Elementary Education'].sum(), df['No. of Inmates benefitted by Adult Education'].sum(), df['No. of Inmates benefitted by Higher Education'].sum(), df['No. of Inmates benefitted by Computer Course'].sum(), df['Total_Benefitted'].sum()]
df

# can also be done by sum(axis =0)
```

```python
#3 Make a bar plot with each state name on the x-axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?
# Make a pie chart that depicts the ratio among different modes of benefits.

# duplicating df and removing 'Total' row
df2 = df.copy()
lenght = len(df2)-1
df2.drop(lenght, axis=0, inplace=True)

plt.figure(figsize=(10,5))

plt.bar(df2['STATE/UT'],df2['Total_Benefitted'], color='green')
plt.title('Statewise Total Benefitted Inmates')
plt.xlabel('State/UT')
plt.ylabel('Total Benefitted Inmates')

plt.show()

print('Andhra Pradesh has the maximum number of beneficiaries')
```

```python
# pie chart 

values = [int(df['No. of Inmates benefitted by Elementary Education'].sum()),
int(df['No. of Inmates benefitted by Adult Education'].sum()),
int(df['No. of Inmates benefitted by Higher Education'].sum()),
int(df['No. of Inmates benefitted by Computer Course'].sum())]

labels = ['No. of Inmates benefitted by Elementary Education',
'No. of Inmates benefitted by Adult Education',
'No. of Inmates benefitted by Higher Education',
'No. of Inmates benefitted by Computer Course']

print(values,labels)

explode = [0.1, 0, 0.2, 0]
plt.title('Ratio of different modes of benefits')
plt.pie(values, labels=labels, autopct='%1.1f%%' ,explode=explode, startangle=90)
plt.legend(labels, loc='best')
plt.show()


# fetch the last row which contains the values of different modes
# x = pd.DataFrame(df.iloc[-1])
# x
```
